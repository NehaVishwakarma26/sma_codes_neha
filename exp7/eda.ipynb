```python
df['likes'].fillna(0, inplace=True)
df['comments'].fillna(0, inplace=True)
df['shares'].fillna(0, inplace=True)

df['text'].fillna("", inplace=True)
```

```python
df['engagement'] = df['likes'] + df['comments'] + df['shares']
```

```python
import matplotlib.pyplot as plt
df['engagement'].plot(kind='hist')
plt.title("Histogram of Engagement")
plt.show()
```

```python
import seaborn as sns
sns.boxplot(x=df['engagement'])
plt.title("Boxplot of Engagement")
plt.show()
```

```python
sns.histplot(df['likes'])
plt.title("Likes Distribution")
plt.show()
```

```python
corr = df[['likes','comments','shares','engagement']].corr()

sns.heatmap(corr, annot=True)
plt.title("Correlation Heatmap")
plt.show()
```
